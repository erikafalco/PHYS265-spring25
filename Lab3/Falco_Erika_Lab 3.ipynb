# Part 1: The Invariant Mass Distribution

```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.integrate import solve_ivp
```

```python
# Loading the Atlas data

data = np.genfromtxt("atlas_z_to_ll.csv", delimiter = ",", skip_header = 1)

# pt = transverse momentum 
pt1_data = data[:,0]
pt2_data = data[:,1]

# eta = n = pseudorapidity
eta1_data = data[:,2]
eta2_data = data[:,3]

# phi = azimuthal angle
phi1_data = data[:,4]
phi2_data = data[:,5]

# E = total energy 
E1_data = data[:,6]
E2_data = data[:,7]
```

```python
# For each lepton, pair I will calculate the mass of a hypothetical particle which decayed to produce that pair. 
# I will first define a function to calculate mass 

def mass(pt1, pt2, eta1, eta2, phi1, phi2, E1, E2):
    
    p1x = pt1*np.cos(phi1)
    p2x = pt2*np.cos(phi2)
    pxtot = p1x + p2x 
    # Total x momentum of both particles  

    p1y = pt1*np.sin(phi1)
    p2y = pt2*np.sin(phi2)
    pytot = p1y + p2y 
    # Total y momentum of both particles 
    
    p1z = pt1*np.sinh(eta1)
    p2z = pt2*np.sinh(eta2)
    pztot = p1z + p2z
    # Total z momentum of both particles

    Etot = E1 + E2
    # Total energy of both particles

    M = np.sqrt(Etot**2 - (pxtot**2 + pytot**2 + pztot**2))
    return M 
    # This function calculates the invariant mass of the boson.

particle_mass = mass(pt1_data, pt2_data, eta1_data, eta2_data, phi1_data, phi2_data, E1_data, E2_data)
# Calculating the boson's invariant mass for each particle pair and storing them in an array
```

```python
# I will now plot the mass distribution in a histogram
fig, ax = plt.subplots();
bins = np.linspace(80, 100, 41);

n_invar_mass, bins, patch = ax.hist(particle_mass, bins = bins, color = "darkblue", label = "Data");

# Assuming a Poisson distribution, I set the errors to be sqrt(number of counts in the bin)
unc_num_events = np.sqrt(n_invar_mass)
bins_midpoint = .5 * (bins[0:-1] + bins[1:])

# Plotting the error bars
ax.errorbar(bins_midpoint, n_invar_mass, unc_num_events, ls = "", color = "cornflowerblue", label = "Poisson error in data", capsize = 2)

ax.set_xlabel(r"Invariant Mass $(GeV/c^2)$")
ax.set_ylabel("Counts")
ax.set_title(r"$Z^0$ Boson Invariant Mass Distribution")

ax.legend()
fig.tight_layout()
```

```python
# I will now define my Breit-Wigner fit function 
def D(m, m0, Gamma):
    num = Gamma/2
    denom = np.pi * ( (m-m0)**2 + (Gamma/2)**2 )
    D = num/denom
    return D

# The normalized function is multiplied by a factor of 5000/2
def D_norm(m, m0, Gamma):
    return (5000/2)* D(m, m0, Gamma)
```

```python
from scipy.optimize import curve_fit

# I am masking the data to only use 87-93 GeV/c^2 to create my fit 
mask = (bins_midpoint > 87) & (bins_midpoint < 93)
bins_masked = bins_midpoint[mask]

# I am now fitting the data 
params, covar = curve_fit(f = D_norm, xdata = bins_masked, ydata = n_invar_mass[mask], sigma = unc_num_events[mask], absolute_sigma = True)

# Here are my fitting parameters 
m0_fit, Gamma_fit = params

# The error in my fitting parameters is calculated using the diagonal components of the covariance matrix 
m0_fit_err = np.sqrt(covar[0][0])
Gamma_fit_err = np.sqrt(covar[1][1])

print(fr"m0 fit value: {m0_fit} +/- {m0_fit_err} GeV/c^2")
print(fr"Gamma fit value: {Gamma_fit} +/- {Gamma_fit_err} GeV/c^2")
```

```python
import scipy.stats as st

# I am now plugging my best fit parameters into the Breit-Wigner function to generate a fitted data set 
n_masses_fit = D_norm(bins_midpoint, m0_fit, Gamma_fit)

# Plotting my original data histogram with errorbars, like before
fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (9, 10), gridspec_kw = {"height_ratios":[3,1], "hspace" : 0.4})
ax1.hist(particle_mass, bins = bins, color = "darkblue", label = "Data");
ax1.errorbar(bins_midpoint, n_invar_mass, unc_num_events, ls = "", color = "cornflowerblue", capsize = 2, label = "Poisson error in data")

# Plotting my fit on top of the original histogram
ax1.plot(bins_midpoint, n_masses_fit, color = "orange", label = "Breit-Wigner fit", ls = "--")
ax1.set_xlabel(r"Invariant mass $(GeV/c^2)$", fontsize = 12)
ax1.set_ylabel("Counts", fontsize = 12)
ax1.set_title(r"$Z^0$ Boson Invariant Mass Distribution", fontsize = 15)

# Establishing where my fit is 
ax1.axvline(87, ls = ":", label = "Fitting range limits")
ax1.axvline(93, ls = ":")

ax1.legend()

n_masses_theory = D_norm(bins_midpoint, m0_fit, Gamma_fit)

# Comparing the theory and the data via a chi-squared test 
chisq = np.sum((n_invar_mass[mask]-n_masses_theory[mask])**2/unc_num_events[mask]**2)
print('Chi squared = ', chisq)

# Calculating the p value and reduced chi-squared
N = len(n_invar_mass[mask])
delta = 2
ndof = N-delta 
pvalue = st.chi2.sf(chisq, ndof)
red_chisq = chisq/ndof
print(f"Reduced Chi squared: {red_chisq}")
print(f"P value = {pvalue}")

####################

# Calculating the residuals (data-thoery) and plotting them on the sub-plot
residuals = n_invar_mass - n_masses_theory
ax2.errorbar(bins_midpoint, residuals, unc_num_events, linestyle = "", marker = "o", color = "cornflowerblue")
ax2.axhline(0, ls = "--", label = "Perfect agreement", color = "orange")
ax2.axvline(87, ls = ":", label = "Fitting range limits")
ax2.axvline(93, ls = ":")
ax2.legend()
ax2.set_xlabel(r"Invariant mass $(GeV/c^2)$", fontsize = 12)
ax2.set_ylabel(r"|Theory - Data| $(GeV/c^2)$", fontsize = 12)
ax2.set_title("Residuals", fontsize = 15)

ax1.text(80, 200, fr"$m_0$ = {m0_fit:.1f} $\pm$ {m0_fit_err:.1f} $GeV/c^2$" + "\n" + fr"$\chi ^2$ / NDOF = {chisq:.1f}/{ndof:.1f}" + f"\nP value = {pvalue:.1f}", fontsize = 12)

fig.suptitle("Figure 1", fontsize = 15, fontweight = "bold");
```

# Part 3: 2D Parameter Contours

```python
# Specifying the bounds of my 2D parameter scan
n_bins = 300
chi_map = np.zeros((n_bins, n_bins))
Mass = np.linspace(89, 91, n_bins)
Width = np.linspace(5, 8, n_bins)

# Iterating over all possible m0 and Gamma combinations within our specified range
# the double for loop 
for i in range(n_bins):
    for j in range(n_bins):
        theory = D_norm(bins_midpoint[mask], Mass[i], Width[j])
        
        chi2 = np.sum((theory-n_invar_mass[mask])**2/unc_num_events[mask]**2)
        chi_map[j, i] = chi2

# Subtracting the minimum chi-squared from the chi-squared at different m0 and Gamma values to calculate delta chi-squared
chi_min = np.min(chi_map)
chi_map = np.clip(chi_map, chi_min, chi_min+35)
delta_chi_map = chi_map - chi_min

# Plotting the delta chi-squared as a color map
X, Y = np.meshgrid(Mass, Width)

fig2, ax2 = plt.subplots(1, 1)
cs = ax2.contourf(X, Y, delta_chi_map, 500, cmap = "Blues_r")
cbar = fig2.colorbar(cs, ax = ax2)
cbar.set_label("Delta chi-square")

# Plotting confidence levels as contours on the color map
levels = [2.30, 9.21]
CS = ax2.contour(X, Y, delta_chi_map, levels = levels, colors = "black", linestyles = ["solid", "dashed"])
ax2.clabel(CS, CS.levels, fmt = {CS.levels[0]: r"$1\sigma$", CS.levels[1]: r"$3\sigma$"}, fontsize = 12)
    
ax2.plot(m0_fit, Gamma_fit, marker = "x", color = "red", label = "Best fit", ls = "")

ax2.set_xlabel(r"$Z^0$ Mass $m_0$ $(GeV/c^2)$", fontsize = 12)
ax2.set_ylabel(r"Width $\Gamma$ $(GeV/c^2)$", fontsize = 12);
ax2.set_title(r"Figure 2: $\Delta  \chi ^2$", fontsize = 15, fontweight = "bold")

ax2.legend()

fig2.tight_layout()
```
