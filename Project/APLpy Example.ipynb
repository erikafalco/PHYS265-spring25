```python
import matplotlib.pyplot as plt 
from astropy.utils.data import get_pkg_data_filename

cube = 'WFPC2u5780205r_c0fx.fits'

import aplpy

fig = plt.figure(figsize = (11, 9.1))

f1 = aplpy.FITSFigure(cube, dimensions = [0, 1], slices=[0], figsize=(8, 6), figure = fig, subplot = [0, 0.6, 0.4, 0.5])
f1.show_colorscale()
f1.add_grid()
f1.tick_labels.set_font(size='medium')
f1.axis_labels.set_font(size='large')
f1.add_colorbar()
f1.colorbar.set_axis_label_text("Counts (DN) per second")
f1.set_title("Slice 1", fontsize = 15)
#######################################################################################
f2 = aplpy.FITSFigure(cube, dimensions = [0, 1], slices=[1], figsize=(8, 6), figure = fig, subplot = [0.6, 0.6, 0.4, 0.5])
f2.show_colorscale()
f2.add_grid()
f2.set_title("Slice 2", fontsize = 15)
f2.tick_labels.set_font(size='medium')
f2.axis_labels.set_font(size='large')
f2.add_colorbar()
f2.colorbar.set_axis_label_text("Counts (DN) per second")
#######################################################################################
f3 = aplpy.FITSFigure(cube, dimensions = [0, 1], slices=[2], figsize=(8, 6), figure = fig, subplot = [0, 0, 0.4, 0.5])
f3.show_colorscale()
f3.add_grid()
f3.set_title("Slice 3", fontsize = 15)
f3.tick_labels.set_font(size='medium')
f3.axis_labels.set_font(size='large')
f3.add_colorbar()
f3.colorbar.set_axis_label_text("Counts (DN) per second")
#######################################################################################
f4 = aplpy.FITSFigure(cube, dimensions = [0, 1], slices=[3], figsize=(8, 6), figure = fig, subplot = [0.6, 0, 0.4, 0.5])
f4.show_colorscale()
f4.add_grid()
f4.set_title("Slice 4", fontsize = 15)
f4.tick_labels.set_font(size='medium')
f4.axis_labels.set_font(size='large')
f4.add_colorbar()
f4.colorbar.set_axis_label_text("Counts (DN) per second")

fig.text(0.5, 1.2, "NGC4151: Seyfert 1 Galaxy\nWFPC II", ha = "center", fontsize = 20, fontweight = "bold")
```
